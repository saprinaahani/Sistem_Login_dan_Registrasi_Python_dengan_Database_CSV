```python
import csv

# Fungsi untuk melakukan login
def login(username, password):
    with open('users.csv', mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row['Nama'] == username and row['Password'] == password:
                return int(row['Hak Akses'])
    return -1  # Return -1 jika login gagal

# Fungsi untuk melakukan registrasi
def register():
    username = input("Masukkan nama pengguna: ")
    password = input("Masukkan kata sandi: ")
    hak_akses = input("Apakah Anda admin? (jawab 'iamgod' jika ya): ")
    
    if hak_akses == 'iamgod':
        hak_akses = 1
    else:
        hak_akses = 0

    with open('users.csv', mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([username, password, hak_akses])
    print("Registrasi berhasil.")

# Fungsi untuk menyambut pengguna
def welcome(user_type):
    with open('hello.txt', 'r') as file:
        hello_message = file.read()

    if user_type == 1:
        hello_message = "Selamat datang, Admin!\n" + hello_message
    else:
        hello_message = "Selamat datang, Tamu!\n" + hello_message

    print(hello_message)

# Fungsi utama
def main():
    print("Selamat datang di program sederhana kami!")

    while True:
        print("\nPilihan:")
        print("1. Login")
        print("2. Registrasi")
        print("3. Keluar")
        choice = input("Pilih opsi: ")

        if choice == '1':
            username = input("Masukkan nama pengguna: ")
            password = input("Masukkan kata sandi: ")
            user_type = login(username, password)
            if user_type == -1:
                print("Login gagal. Coba lagi atau registrasi.")
            else:
                welcome(user_type)
        elif choice == '2':
            register()
        elif choice == '3':
            print("Terima kasih telah menggunakan program kami.")
            break
        else:
            print("Opsi tidak valid. Silakan pilih lagi.")

if __name__ == "__main__":
    main()
```
